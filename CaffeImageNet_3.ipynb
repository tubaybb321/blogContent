```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import sys
import random

caffe_home='/home/me/Caffe/'
caffe_image_dir = os.path.join(caffe_home,'images')

sys.path.insert(0, caffe_home + 'caffe/python')
import caffe

# Set the right path to your model definition file, pretrained model weights,
# and the image you would like to classify.
MODEL_FILE = os.path.join(caffe_home,'caffe/models/bvlc_reference_caffenet/deploy.prototxt')
PRETRAINED = os.path.join(caffe_home,'caffe/models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel')
```

```python
import pickle
classDictFile = open(os.path.join(caffe_home,'caffe/data/ilsvrc12/classDict.pkl'),'rb')
classDict = pickle.load(classDictFile)
nLabelDictFile = open(os.path.join(caffe_home,'caffe/data/ilsvrc12/nLabelDict.pkl'),'rb')
nLabelDict = pickle.load(nLabelDictFile)
```

```python
caffe.set_mode_gpu()
net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       mean=np.load(caffe_home + 'caffe/python/caffe/imagenet/ilsvrc_2012_mean.npy').mean(1).mean(1),
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))
```

```python
ctr = 0
SRC_DIR = '/media/me/OSDisk/ILSVRC2012/ILSVRC2012_img_train'
```

```python
if ctr == 0:
  random.seed(10)
imageDirs = [x for x in os.listdir(SRC_DIR) if os.path.isdir(os.path.join(SRC_DIR,x))]
currImDir = imageDirs[random.randrange(len(imageDirs))]
images = os.listdir(os.path.join(SRC_DIR,currImDir))
IMAGE = images[random.randrange(len(images))]
IMAGE_FILE = os.path.join(SRC_DIR, currImDir, IMAGE)
GND_TRUTH = nLabelDict[IMAGE.split('_')[0].lower()]
input_image = caffe.io.load_image(IMAGE_FILE)
plt.imshow(input_image)
prediction = net.predict([input_image]) 
predSort =np.argsort(prediction)[0][::-1]
print "Image File:", IMAGE_FILE
print "Ground Truth:", GND_TRUTH
for i in range(10):
    print i, classDict[predSort[i]][1], prediction[0][predSort[i]]
ctr += 1
```
